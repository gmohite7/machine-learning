#### Normalization

It is a technique often applied as part of data preparation for machine learning.The goal of normalization is
to change the values of numeric columns in the dataset to a use a common scale,without distorting differences in the range of values 
or loosing information

e.g using heaight and weight predict if the person buys equipment or not.Always eliminate unit(kg,pounds etc)

Types of Normalization

1. Min max normalisation : o/p ==> 0,1 (very commonly used)
2. Mean normalisation : o/p ==> -1,1 (very rarely used)
3. Max absolute scaling : MaxAbsScaler : Mostly used when you have sparse data
4. Robust scaling : Robust to outliers (Generally performs well when you have lot of outliers)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns = ['Class label','Alcohol','Malic acid']
```

```python
df
```

```python
# distribution plot
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict = {1:'red',2:'blue',3:'green'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.drop('Class label',axis=1),
                                                df['Class label'],
                                                test_size=0.3,
                                                random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# fit the scaler to the train set ,it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(x=X_train['Alcohol'],y=X_train['Malic acid'],c=y_train)
ax1.set_title('Before scaling')

ax2.scatter(x=X_train_scaled['Alcohol'],y=X_train_scaled['Malic acid'],c=y_train)
ax2.set_title('After scaling')
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

# before scaling
ax1.set_title('Before scaling')
sns.kdeplot(X_train['Alcohol'],ax=ax1)
sns.kdeplot(X_train['Malic acid'],ax=ax1)

# after scaling
ax2.set_title('After scaling')
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

# before scaling
ax1.set_title('Alcohol distribution before scaling')
sns.kdeplot(X_train['Alcohol'],ax=ax1)

#After scaling
ax2.set_title('Alcohol distribution after standard scaling')
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)

plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize = (12,5))

#before scaling
ax1.set_title('Malic acid distribution before scaling')
sns.kdeplot(X_train['Malic acid'],ax=ax1)

#After scaling
ax2.set_title('Malic acid distribution after scaling')
sns.kdeplot(X_train_scaled['Malic acid'],ax=ax2)

plt.show()
```
