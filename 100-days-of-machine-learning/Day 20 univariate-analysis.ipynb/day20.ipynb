```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

#### 1. Categorical data

a. Countplot

```python
sns.countplot(df, x="Embarked", hue="Embarked")
```

#### b. Piechart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

#### 2. Numerical data

#### a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

#### b. Distplot

```python
sns.distplot(df['Age'])
```

#### c. Boxplot

```python
sns.boxplot(x=df['Age'])
```
